# Linked list

```python
class Node:
    def __init__(self,data,next):
        self.data = data
        self.next = next
        
class LinkedList:
    def __init__(self):
        self.head = None
        
    def insert_at_beggining(self,data):
        node = Node(data,self.head)
        self.head = node
        
    def print(self):
        if self.head is None:
            print("Linked list is empty.")
            return 
        else:
            itr = self.head
            llstr = " "
            
            while itr:
                llstr += str(itr.data) + "-->"
                itr = itr.next
            print(llstr)
            
    def insert_at_end(self,data):
        if self.head is None:
            self.head = Node(data,None)
            return
        else:
            itr = self.head
            while itr.next:
                itr = itr.next
            
        itr.next = Node(data,None)
        
    def insert_values(self,data_list):
        self.next = None
        for data in data_list:
            self.insert_at_end(data)
            
    def get_length(self):
        count = 0
        itr = self.head
        while itr:
            count += 1
            itr = itr.next
        
        return count
    
    def remove_at(self,index):
        if index<0 or index>self.get_length():
            raise Exception("Invalid index")
        elif index == 0:
            self.head = self.head.next
            return
        else:
            count = 0
            itr = self.head
            while itr:
                if count ==index-1:
                    itr.next = itr.next.next
                    break
                
                itr = itr.next
                count +=1
                
    def insert_at(self,index,data):
        if index<0 or index>self.get_length():
            raise Exception("Invalid index")
        elif index==0:
            self.insert_at_beggining(data)
            return
        else:
            count = 0
            itr = self.head
            while itr:
                if count == index-1:
                    node = Node(data,itr.next)
                    itr.next = node
                    break
                else:
                    itr = itr.next
                    count +=1
                    
    def insert_after_value(self,data_after,data_to_insert):
        if self.head is None:
            return
        elif self.head.data == Node(data_to_insert,self.head.next):
            return
        else:
            itr = self.head
            while itr:
                if itr.data == data_after:
                    itr.next = Node(data_to_insert,itr.next)
                    break
                else:
                    itr = itr.next
                    
    def remove_by_value(self,data):
        if self.head is None:
            return
        elif self.head.data == data:
            self.head = self.head.next
            return
        else:
            itr = self.head
            while itr.next:
                if itr.next.data == data:
                    itr.next = itr.next.next
                    break
                else:
                    itr = itr.next
        
               
if __name__=="__main__":
    ll = LinkedList()
    ll.insert_at_beggining(32)
    ll.print()
    ll.insert_at_end(53)
    ll.print()
    ll.insert_values(['banana','mango','grapes','orange'])
    ll.print()
    ll.insert_at_beggining(76)
    ll.print()
    ll.insert_at_end(98)
    ll.print()
    print("Length:",ll.get_length())
    ll.print()
    ll.remove_at(2)
    ll.print()
    ll.insert_at(4,"figs")
    ll.print()
    ll.insert_after_value("mango","guava")
    ll.print()
    ll.remove_by_value("orange")
    ll.print()
```
